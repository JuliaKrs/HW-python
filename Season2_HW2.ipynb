```python
 
def get_data(features, predicted):
    
    gender_code = {"gender": {"Male" : 1, "Female" : 0}}
    newdata = dfs.replace(gender_code)
    
    now = pd.Timestamp('now')
    newdata['birthdate'] = pd.to_datetime(dfs['birthdate'], format='%Y-%m-%d')
    newdata["BMI"] = newdata["weight"] / (newdata["height"]**2)
    newdata = newdata.sample(frac=1)
    
    global allinput, alloutput
    allinput = newdata[features]
    alloutput = newdata[predicted]
    print(allinput, alloutput)    
    
def sport_data(features, predicted_data):
    predicted_gender = {'gender':{'Male':0, 'Femaale':1}}

def train_test_model():
    global X_train, X_test, Y_train, Y_test
    X_train, X_test, Y_train, Y_test = model_selection.train_test_split(allinput, alloutput, test_size=0.02)
    
    global model
    model = linear_model.LogisticRegression()
    model.fit(X_train, Y_train)
    
    global Y_test_predicted
    Y_test_predicted = model.predict(X_test)


    
def accuracy():
    [tn, fp], [fn, tp] = confusion_matrix(Y_test, Y_test_predicted)
    confusion_matrix(Y_test, Y_test_predicted)
    
    precision = tp / (tp + fp)
    recall = tp / tp + fn

    print("\n\n Precision: ", tp / (tp + fp))
    print("\n Recall: ", tp / (tp + fn))
    print("\n Accuracy: ", (tn + tp) / (tp + fp + tn + fn))
    print("\n F1: ",  (precision * recall) / ((precision) + recall))
    
    report = classification_report(Y_test, model.predict(X_test), target_names=['Male', 'Female'])
    print(report)
    
```
